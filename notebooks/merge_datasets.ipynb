```python
import pandas as pd


# Load CSV files
games = pd.read_csv("../cleaned_data/games_cleaned.csv")
sales = pd.read_csv("../cleaned_data/vgsales_cleaned.csv")

pd.set_option('display.max_columns', 150)
```

```python
games.head()
```

```python
sales.head()
```

```python
import re
```

```python
def clean_key(s):
    s = re.sub(r"[:\-\.,]", "", s)
    return s.lower().strip()
```

```python
games['merge_key']=games['Title'].apply(clean_key)
sales['merge_key']=sales['Name'].apply(clean_key)
```

```python
games.head()
```

```python
sales = sales.drop(columns=['Name'])
```

```python
merged = pd.merge(
    games,
    sales,
    how='left',
    on = 'merge_key'
)
```

```python
merged.head()
```

```python
merged = merged.drop(columns=['merge_key'])
```
