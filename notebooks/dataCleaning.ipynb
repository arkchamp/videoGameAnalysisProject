```python
vgsales_df['Year']=vgsales_df['Year'].astype(int)
```

```python
games_df['Release Date'] = pd.to_datetime(games_df['Release Date'], errors='coerce')
```

```python
games_df['Release Date'].isnull().sum()
```

```python
games_df.dropna(subset=['Release Date'], inplace=True)
```

```python
def convert_k_to_number(val):
    if isinstance(val, str):
        val = val.replace(',','').strip()
        if 'K' in val:
            return int(float(val.replace('K',''))*1000)
        elif 'M' in val:
            return int(float(val.replace('M',''))*100000) #expecting if m is there or shows up
        else:
            return int(float(val))
    return val
```

```python
cols_to_convert = ['Plays', 'Playing', 'Backlogs', 'Wishlist', 'Times Listed', 'Number of Reviews']
for col in cols_to_convert:
    games_df[col] = games_df[col].apply(convert_k_to_number) #.apply
```

```python
games_df.head(10)
```

```python
def clean_column(val):
    val = val.replace("[", "").replace("]", "").replace("'", "")
    items = [item.strip() for item in val.split(",")]   #splits every words bewtween , and strips to remove white spaces
    return ", ".join(items)
```

```python
games_df['Genres'] = games_df['Genres'].apply(clean_column)
games_df['Team'] = games_df['Team'].apply(clean_column)
```

```python
if 'Unnamed: 0' in games_df.columns:
    games_df.drop(columns=['Unnamed: 0'], inplace=True)
```
